¿Qué significa la Asimetría Positiva?

Significa que la mayoría de los datos son valores pequeños, pero existen algunos pocos valores muy altos que "llevan" el promedio hacia la derecha.

**Reflexión antes del test:**

A simple vista, los clientes con Churn=Yes parecen tener menos antigüedad. Pero necesitamos demostrarlo con un test estadístico.

**¿Por qué NO usamos el test t de Student aquí?**

El test t asume que los datos siguen una **distribución normal**. Veamos si eso se cumple:

```python
# Calcular asimetría (skewness) de las variables numéricas
# Regla: si |skewness| > 0.5 → la distribución NO es simétrica → no usar t-Student

print('Asimetría (skewness) de variables numéricas:')
print('─' * 45)

for col in ['tenure', 'MonthlyCharges', 'TotalCharges']:
    skew = df[col].skew()
    normal = '✓ Simétrica' if abs(skew) < 0.5 else '✗ Asimétrica → usar Mann-Whitney'
    print(f'  {col:<20}  skewness = {skew:+.3f}   {normal}')
```

Este bloque:
- Calcula la asimetría de cada variable numérica.
- Decide si la distribución es:
- Simétrica (normal) → se podría usar t-test
- Asimétrica → mejor usar Mann-Whitney

TotalCharges tiene valores muy altos que llevan la distribución hacia la derecha
Por eso NO conviene usar t-test
Se usa Mann-Whitney U, que no necesita normalidad

**Pregunta 1:** ¿Qué significa que una variable tenga asimetría positiva (skewness > 0)? ¿Qué dice eso sobre la forma de su distribución?

**Respuesta 1:** La asimetría positiva significa que la mayoría de los valores son bajos, pero existen algunos valores muy altos que alargan la distribución hacia la derecha.

## Paso 2: Test de Mann-Whitney U — Variables numéricas

### ¿Qué es el test Mann-Whitney U?

Es un test **no paramétrico**: no requiere que los datos sean normales. Compara si la distribución de una variable es diferente entre **dos grupos independientes**.

| Elemento | Descripción |
|---|---|
| **H₀ (hipótesis nula)** | La distribución de la variable es IGUAL en ambos grupos |
| **H₁ (hipótesis alternativa)** | Existen diferencias entre los grupos |
| **p-value < 0.05** | Rechazamos H₀ → la diferencia es estadísticamente real |
| **p-value ≥ 0.05** | No podemos rechazar H₀ → no hay evidencia suficiente |

> 💡 **Para entenderlo:** Si el p-value es 0.001, significa que la probabilidad de observar esta diferencia por puro azar es del 0.1%. Casi imposible. La diferencia es real.

```python
import scipy.stats as stats  # Librería para pruebas estadísticas

# Separar los dos grupos según Churn
grupo_churn_si = df[df['Churn'] == 'Yes']  # Clientes que cancelaron
grupo_churn_no = df[df['Churn'] == 'No']   # Clientes que NO cancelaron

print('TEST DE MANN-WHITNEY U')
print('H₀: La distribución es igual en ambos grupos')
print('Nivel de significancia: α = 0.05')
print('─' * 70)

# Encabezado de la tabla
print(f'{"Variable":<20} {"Media Churn=Sí":>15} {"Media Churn=No":>15} {"p-value":>12} {"Conclusión"}')
print('─' * 70)

resultados_mw = {}  # Diccionario para guardar p-values

# Evaluar cada variable numérica
for col in ['tenure', 'MonthlyCharges', 'TotalCharges']:
    g1 = grupo_churn_si[col].dropna()  # Datos sin nulos (Churn=Yes)
    g2 = grupo_churn_no[col].dropna()  # Datos sin nulos (Churn=No)

    # Aplicar test Mann-Whitney
    stat, pval = stats.mannwhitneyu(g1, g2, alternative='two-sided')

    resultados_mw[col] = pval  # Guardar resultado

    # Interpretar el resultado
    conclusion = '✓ DIFERENCIA REAL' if pval < 0.05 else '✗ Sin diferencia'

    # Mostrar resultados
    print(f'{col:<20} {g1.mean():>15.2f} {g2.mean():>15.2f} {pval:>12.6f}  {conclusion}')
```

    La hipótesis nula (H₀) indica que no hay diferencias entre los grupos. Por ejemplo, que los clientes que cancelan y los que no cancelan tienen el mismo comportamiento en una variable como MonthlyCharges.
    La hipótesis alternativa (H₁) indica que sí hay diferencias entre los grupos.

*Observación:*

- Todos los p-values son: p = 0.000000

Por tanto:
- p < 0.05 → rechazamos H₀
- Hay diferencias reales entre los grupos
