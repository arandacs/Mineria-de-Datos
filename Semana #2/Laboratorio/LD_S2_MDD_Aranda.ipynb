| N° | Concepto / Principio | Definición con sus propias palabras |
|:---:|---|---|
| 1 | CRISP-DM (definición y significado de las siglas) | CRISP-DM (Cross-Industry Standard Process for Data Mining) es una metodología completa que no solo dice cómo analizar datos, sino cómo llevar todo el proyecto desde entender el problema hasta aplicar la solución en la vida real. Incluye tanto el lado técnico como el del negocio. Organiza un proyecto de análisis de datos en 6 fases secuenciales e iterativas.|
| 2 | SEMMA (definición y significado de las siglas) | SEMMA (Sample, Explore, Modify, Model, Assess). Es una metodología desarrollada por SAS que describe el proceso técnico de construcción de modelos analíticos, partiendo desde la selección de una muestra representativa de datos hasta la evaluación del modelo resultante. Es una metodología más enfocada en los datos. |
| 3 | Comprensión del negocio (Business Understanding) |Se define qué problema se quiere resolver y por qué es importante. Esta etapa es la base, se plantea el objetivo y se entiende la necesidad real.|
| 4 | Comprensión de los datos (Data Understanding) |Se revisan los datos para entender qué contienen, su calidad, y si llegaran a servir. Es importante porque permite detectar errores y problemas antes de avanzar.|
| 5 | Preparación de datos (Data Preparation) |Se dejan los datos listos para usar. Se limpian, se ordenan y se seleccionan. Una de las fases más importantes porque un modelo depende de datos bien preparados.|
| 6 | Modelado (Modeling) |Se aplican algoritmos para encontrar patrones o hacer predicciones. Es donde se construye la solución, pero depende mucho de lo bien que se prepararon los datos antes.|
| 7 | Evaluación (Evaluation) |Se revisa si el modelo funciona bien y si realmente soluciona el problema inicial. Evita usar modelos que dan resultados engañosos. Se verifica si los resultados son confiables.|
| 8 | Despliegue (Deployment) |Cuando el modelo se usa y aplica en la vida real. Esta fase conecta el análisis con la toma de decisiones.|
| 9 | ¿Cuál es la principal diferencia entre CRISP-DM y SEMMA? |CRISP-DM abarca todo el proyecto, incluyendo entender el problema del negocio, mientras que SEMMA se enfoca más en trabajar directamente con los datos y construir modelos.|
| 10 | ¿Por qué CRISP-DM es un modelo cíclico e iterativo? |Porque el proceso no es lineal, este permite volver a fases anteriores para ajustar o mejorar el trabajo realizado. Esto asegura que los resultados sean más precisos y alineados con el problema planteado.|

## 🔹 Paso 3: Fase 3 — Preparación de Datos (Data Preparation)

Ejecuta el siguiente pipeline de preparación de datos. Esta fase consume entre el **60% y 80%** del tiempo de un proyecto real.

*Referencia: Provost, F. y Fawcett, T. (2013). Data Science for Business. O'Reilly Media. Cap. 2.*



> Aquí se dejan los datos listos para el modelado. Se corrigen errores, se transforman variables y se eliminan columnas que no aportan al modelo.



```python
print('=' * 60)
print('FASE 3: PREPARACIÓN DE DATOS')
print('=' * 60)

# === LIMPIEZA ===
# Corregir TotalCharges (contiene espacios en blanco en lugar de valores)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce') # Convierte TotalCharges a número, lo inválido pasa a NaN
print(f'\nValores nulos en TotalCharges después de conversión: {df["TotalCharges"].isnull().sum()}')

# Imputar nulos con la mediana
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
print(f'Valores nulos después de imputación: {df["TotalCharges"].isnull().sum()}')
print('\n✅ Limpieza completada')
```

```python
# === TRANSFORMACIÓN ===
# Codificar variable objetivo
df['Churn_num'] = df['Churn'].map({'Yes': 1, 'No': 0}) # Convierte Churn a valores numéricos
print('Variable objetivo codificada: Churn_num (Yes=1, No=0)')

# Eliminar columna customerID (no aporta al modelo)
df_model = df.drop(columns=['customerID', 'Churn'])  # Elimina identificador y variable textual objetivo
print(f'\nColumna customerID eliminada') # Muestra el tamaño final del dataset
print(f'Dimensiones actuales: {df_model.shape}') # Muestra algunas columnas generadas
```

```python
# === CODIFICACIÓN ===
# Codificar variables categóricas con One-Hot Encoding
# separar variables numéricas y categóricas
num_cols = df_model.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df_model.select_dtypes(include=['object']).columns

# One-Hot Encoding solo para categóricas
df_encoded = pd.get_dummies(df_model, columns=cat_cols, drop_first=True)

print(f'Dimensiones después de One-Hot Encoding: {df_encoded.shape}')
print(f'\nPrimeras 10 columnas del dataset codificado:')
print(df_encoded.columns[:10].tolist())
print(f'\nÚltimas 10 columnas:')
print(df_encoded.columns[-10:].tolist())

print(f'\n✅ Preparación de datos completada')
print(f'Dataset listo para modelado: {df_encoded.shape[0]} registros × {df_encoded.shape[1]} variables')
```

## 🔹 Paso 4: Fase 4 — Modelado (Modeling)

Aplica un modelo de clasificación para predecir el churn. Usaremos un **árbol de decisión** como primer modelo.

*Referencia: Provost, F. y Fawcett, T. (2013). Data Science for Business. O'Reilly Media. Cap. 3.*



> Se construye el primer modelo predictivo. Se usa un árbol de decisión porque es fácil de interpretar y sirve como punto de partida para comparar después con otros modelos más avanzados.



```python
from sklearn.model_selection import train_test_split  # Divide los datos en entrenamiento y prueba
from sklearn.tree import DecisionTreeClassifier  # Modelo de árbol de decisión
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix  # Métricas de evaluación

print('=' * 60)
print('FASE 4: MODELADO')
print('=' * 60)

# Separar features (X) y target (y)
X = df_encoded.drop(columns=['Churn_num'])  # Variables predictoras
y = df_encoded['Churn_num']  # Variable objetivo

print(f'\nFeatures (X): {X.shape}')
print(f'Target (y): {y.shape}')
print(f'\nDistribución del target:')
print(y.value_counts())
```

```python
# Dividir en entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Muestra cuántos datos van a entrenamiento
print(f'Conjunto de entrenamiento: {X_train.shape[0]} registros ({X_train.shape[0]/len(X)*100:.0f}%)')
# Muestra cuántos datos van a prueba
print(f'Conjunto de prueba: {X_test.shape[0]} registros ({X_test.shape[0]/len(X)*100:.0f}%)')

print(f'\nDistribución en entrenamiento:')
print(y_train.value_counts(normalize=True) * 100)
print(f'\nDistribución en prueba:')
print(y_test.value_counts(normalize=True) * 100)
```

```python
# Entrenar árbol de decisión
modelo_arbol = DecisionTreeClassifier(max_depth=5, random_state=42) # Árbol con profundidad máxima 5
modelo_arbol.fit(X_train, y_train) # Entrena el modelo

# Predicciones
y_pred = modelo_arbol.predict(X_test) # Genera predicciones sobre el conjunto de prueba

print('✅ Modelo entrenado exitosamente')
print(f'\nParámetros del modelo:')
print(f'  Algoritmo: Árbol de Decisión')
print(f'  Profundidad máxima: {modelo_arbol.max_depth}')
print(f'  Nodos en el árbol: {modelo_arbol.tree_.node_count}')
print(f'  Features utilizados: {modelo_arbol.n_features_in_}')
```

### ✏️ Pregunta 8

**Respuesta:**
*Dividimos los datos en entrenamiento y prueba para evaluar el modelo de manera objetiva y honesta con datos que no ha visto previamente durante el aprendizaje. De lo contrario no sería tan preciso y solo estaría memorizando.*

*En este caso, se utilizó 70% de los datos para entrenamiento (4930 registros) y 30% para prueba (2113 registros).*

### ✏️ Pregunta 9
**¿Qué significa el parámetro `max_depth=5` en el árbol de decisión? ¿Qué pasaría si no lo limitamos?**

**Respuesta:**

*El parámetro max_depth=5 limita y controla la profundidad del árbol de decisión a cinco niveles, evitando que el modelo se vuelva demasiado complejo.*

*Si no se limitara, el árbol podría crecer en exceso y ajustarse demasiado a los datos de entrenamiento (sobreajuste), lo que reduciría su capacidad para generalizar a nuevos datos.*

*"Con menos nodos el modelo es más simple, mientras que con más nodos es más complejo."*

## 🔹 Paso 5: Fase 5 — Evaluación (Evaluation)

Evalúa el rendimiento del modelo con métricas de clasificación. No basta con que funcione técnicamente: debe cumplir los **objetivos del negocio** definidos en la Fase 1.

*Referencia: Provost, F. y Fawcett, T. (2013). Data Science for Business. O'Reilly Media. Cap. 3.*



> Esta fase sirve para saber si el modelo realmente funciona. Hay que medirlo con métricas y revisar en qué casos se equivoca.



```python
print('=' * 60)
print('FASE 5: EVALUACIÓN DEL MODELO')
print('=' * 60)

# Accuracy
accuracy = accuracy_score(y_test, y_pred) # Calcula el porcentaje de aciertos
print(f'\n📊 Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)')
```

> Un accuracy de 78.8% parece bueno, pero no es suficiente para decir que el modelo es recomendable. Porque el dataset está desbalanceado (≈73.5% “No churn” y ≈27% “Churn”). Por ello es escencial evaluar el recall.

> El modelo tiene recall de churn ≈ 0.52, lo que significa que solo detecta alrededor del 52% de los clientes que realmente se van.

```python
# Reporte de clasificación completo
print('Reporte de clasificación:')
print('=' * 55)
print(classification_report(y_test, y_pred, target_names=['No Churn (0)', 'Churn (1)']))  # Precision, recall y F1
```

El modelo muestra un desempeño mejor en la clase “No Churn” que en “Churn”.

- Para No Churn (0):
Tiene precision 0.84 y recall 0.89, lo que indica que el modelo identifica correctamente la mayoría de clientes que no cancelan y además se equivoca poco al clasificarlos.

- Para Churn (1):
Tiene precision 0.62 y **recall 0.52**, lo que significa que cuando predice churn, acierta de forma moderada (62%).
Pero solo detecta el 52% de los clientes que realmente se van.
Esto evidencia una debilidad importante en la clase más relevante.





> El modelo presenta un buen desempeño en la predicción de clientes que no cancelan, pero tiene un rendimiento limitado en la detección de churn, especialmente por su bajo recall (0.52). Esto indica que no logra identificar a una parte importante de los clientes en riesgo, lo cual es una debilidad importante considerando el objetivo del negocio que es anticipar el churn.


```python
# Matriz de confusión visual
cm = confusion_matrix(y_test, y_pred) # Calcula aciertos y errores por clase

fig, ax = plt.subplots(figsize=(7, 5)) # Crea la figura
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Churn', 'Churn'],
            yticklabels=['No Churn', 'Churn'], ax=ax,
            annot_kws={'size': 16})
ax.set_title('Matriz de Confusión — Árbol de Decisión', fontsize=14, fontweight='bold')
ax.set_xlabel('Predicción', fontsize=12)
ax.set_ylabel('Valor Real', fontsize=12)
plt.tight_layout()
plt.show()

# Interpretación numérica
print(f'Verdaderos Negativos (TN): {cm[0][0]} — Clientes que NO cancelaron, predichos correctamente')
print(f'Falsos Positivos (FP):     {cm[0][1]} — Clientes que NO cancelaron, pero el modelo dijo que sí')
print(f'Falsos Negativos (FN):     {cm[1][0]} — Clientes que SÍ cancelaron, pero el modelo NO los detectó')
print(f'Verdaderos Positivos (TP): {cm[1][1]} — Clientes que SÍ cancelaron, detectados correctamente')
```
